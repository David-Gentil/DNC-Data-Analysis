```python
#Importando Biblioteca Pandas

import pandas as pd
```

```python
aluguel_casa = pd.read_excel('/content/houses_rent.xlsx')
```

```python
aluguel_casa.head()
```

```python
aluguel_casa.tail()
```

```python
aluguel_casa = aluguel_casa.rename(columns={'parking spaces':'parking_spaces', 'rent amount':'rent_amount', 'property tax':'property_tax', 'fire insurance':'fire_insurance'})
aluguel_casa.head()
```

```python
aluguel_casa.dtypes
```

```python
aluguel_casa = aluguel_casa.astype({'rent_amount':'float', 'property_tax':'float', 'fire_insurance':'float', 'total':'float'})
aluguel_casa.dtypes
```

```python
pd.options.display.float_format = '{:,.2f}'.format
aluguel_casa.head()
```

```python
aluguel_casa.info()
```

```python
aluguel_casa[['area', 'rooms', 'bathroom', 'parking_spaces', 'rent_amount', 'property_tax', 'fire_insurance', 'total']].describe()
```

```python
#Fazendo filtro de valor <= 4000

aluguel_casa = aluguel_casa[(aluguel_casa['total']<=4000)]
aluguel_casa.head()
```

```python
aluguel_casa.describe()
```

```python
#Fazendo filtro de valor <= 4000 e que aceite animais

aluguel_casa = aluguel_casa[(aluguel_casa['total']<=4000) & (aluguel_casa['animal']=='accept')]
aluguel_casa.head(10)
```

```python
#Fazendo filtro de valor <= 4000 e que aceite animais ou

aluguel_casa = aluguel_casa[(aluguel_casa['total']<=4000) & (aluguel_casa['animal']=='accept') & ((aluguel_casa['rooms']==2)|(aluguel_casa['rooms']==3))]
aluguel_casa.head(10)
```

```python
aluguel_casa.describe()
```

```python
aluguel_casa.query('total <= 4000 & animal == "accept" & (rooms == 2 | rooms == 3)')
```

```python
#Agrupando cidade por preo mdio

agg_city_price = aluguel_casa.groupby(['city']).agg({'total':'mean'}).reset_index()
agg_city_price
```

```python
#Agrupando nmero de banheiros e quartos por pero mdio

agg_rooms_bathroom = aluguel_casa.groupby(['rooms', 'bathroom']).agg({'total':'mean'}).reset_index()
agg_rooms_bathroom
```
