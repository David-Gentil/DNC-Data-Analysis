```python
#Importando Biblioteca Pandas

import pandas as pd
```

```python
#Importando Biblioteca Seaborn

import seaborn as sns
```

```python
#Agrupando número de banheiros e quartos por perço médio
aluguel_casa = pd.read_excel('/content/houses_rent.xlsx')
aluguel_casa = aluguel_casa.query('total <= 4000 & animal == "accept" & (rooms == 2 | rooms == 3)')
```

```python
agg_rooms_bathroom = aluguel_casa.groupby(['rooms', 'bathroom']).agg({'total':'mean'}).reset_index()
agg_rooms_bathroom
```

```python
ax = sns.barplot(x='rooms', y='total', data=agg_rooms_bathroom);
ax.set_title('Preço médio por quartos');
ax.set_ylabel('Preço médio');
ax.set_xlabel('Quantidade de quartos');
```

```python
ax = sns.barplot(x='bathroom', y='total', data=agg_rooms_bathroom, color = 'purple');
ax.set_title('Preço médio por banheiros');
ax.set_ylabel('Preço médio')
ax.set_xlabel('Quantidade de banheiros');
```
